```python
import pickle
import pandas as pd
import scipy.sparse as sp
```

```python
# Load model
with open("RiskEnsembleClassifier.pkl", "rb") as f:
    bundle = pickle.load(f)

model   = bundle["model"]
tfidf_w = bundle["tfidf_word"]
tfidf_c = bundle["tfidf_char"]
scaler  = bundle["scaler"]
le      = bundle["label_encoder"]
num_cols = bundle["num_cols"]
```

```python
df_test = pd.DataFrame(test_data)
df_test.head()
```

```python
# Transform
word = tfidf_w.transform(df_test["clean_text"])
char = tfidf_c.transform(df_test["clean_text"])
num  = sp.csr_matrix(scaler.transform(df_test[num_cols]))
X    = sp.hstack([word, char, num], format="csr")
```

```python
# Predict
preds = le.inverse_transform(model.predict(X))

for i, (pred, sample) in enumerate(zip(preds, test_data)):
    print(f"Sample {i+1}: {pred:10s} | {sample['clean_text'][:60]}...")
```
